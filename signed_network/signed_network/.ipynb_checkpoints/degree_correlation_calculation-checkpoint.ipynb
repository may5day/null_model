```python
def loadfile():
    with open(filename) as f:
        edges = []
        for each_line in f:
            #print each_line
            n1,n2,w = each_line.strip('\t\r\n').split('\t')
            edges.append((n1,n2,int(w)))
        return edges
```

```python
#filename = 'N46edge.txt'
filename = 'epinions.txt'
G = nx.Graph()
G.add_weighted_edges_from(loadfile())
```

```python
nswap=200000
max_tries=500000
```

```python
G1 = sm.sign_network_positive_swap(G, nswap, max_tries)           #正边交换
G2 = sm.sign_network_negative_swap(G, nswap, max_tries)           #负边交换
G3 = sm.sign_network_negative_swap(G1, nswap, max_tries)          #正负边分别进行交换：先交换正边，再交换负边
G4 = sm.sign_network_full_swap(G, nswap, max_tries)               #正负边同时进行交换
G5 = sm.sign_network_sign_swap(G, nswap, max_tries)               #符号交换
```

```python
def divide_network(G):
    edge_list=[]
    for i,j,weight_data in G.edges(data=True):
        if 'weight' in weight_data:
            edge_list.append([i,j,weight_data['weight']])
            
    positive_edges=[]
    negitive_edges=[]
    for item in edge_list:
        if item[2]==1:       
            positive_edges.append(item)
        if item[2]==2:       
            negitive_edges.append(item)
        
    Gp= nx.Graph()
    Gp.add_weighted_edges_from(positive_edges)
        
    Gn= nx.Graph()
    Gn.add_weighted_edges_from(negitive_edges)
    return Gp,Gn
```

```python
Gp,Gn=divide_network(G)
rp0=nx.degree_assortativity_coefficient(Gp)
rn0=nx.degree_assortativity_coefficient(Gn)
print("The assortativity coefficient of positive edges in the original network is ","%3.3f" %rp0)
print("The assortativity coefficient of negative edges in the original network is ","%3.3f" %rn0)
```

```python
def knn_to_list(knn_dict):
    list_keys=list(knn_dict)
    knn_list=[]
    for item in list_keys:
        knn_list.append((int(item),knn_dict[item]))
    knn_list_new=sorted(knn_list, key=lambda knn_list: knn_list[0])
    
    knn_degrees=[]
    knn_values=[]
    for item in knn_list_new:
        knn_degrees.append(item[0])
        knn_values.append(item[1])
    
    return knn_degrees,knn_values
```

```python
def CommonNeighbor(G,nodeij):
    node_i=nodeij[0]
    node_j=nodeij[1]
    neigh_i=set(G.neighbors(node_i))
    neigh_j=set(G.neighbors(node_j))
    neigh_ij=neigh_i.intersection(neigh_j)
    num_cn=len(neigh_ij)       
    return num_cn
```

```python
def CN_embedding(G,Gpn):
    CN_list=[]
    for item in G.edges():
        try:
            item_CN=CommonNeighbor(Gpn,item)
            CN_list.append((item_CN,G[item[0]][item[1]]['weight']))
        except:
            pass

    CN_list_sort=sorted(CN_list, key=lambda CN_list: CN_list[0])
    degree_list=[]
    positive_edge_cn=[]

    positve_edge_number=0
    sum_edge_number=0
    item_index=0
    for item in CN_list_sort:
        if item[0]==item_index:
            if item[1]==1:
                positve_edge_number=positve_edge_number+1
            sum_edge_number=sum_edge_number+1
        elif item[0]>item_index:
            if sum_edge_number==0:
                degree_list.append(item_index)
                positive_edge_cn.append(0)
            else:  
                positive_edge_cn.append([item_index,float(positve_edge_number)/sum_edge_number])
            item_index=item[0]
            positve_edge_number=0
            sum_edge_number=0
    return positive_edge_cn
```

```python
Gp,Gn=divide_network(G)
positive_edge_cn_p=CN_embedding(G,Gp)
positive_edge_cn_n=CN_embedding(G,Gn)
```
